```python
import numpy as np 
import matplotlib.pyplot as plt 
import string
from urllib.request import urlopen as url
import re 
```

### Reading and Storing 

```python
# Part 1. Reading data from files 

link = 'http://www.gutenberg.org/files/2701/2701-0.txt'

for line in url(link):
    line = line.decode().rstrip().lower()
    print(line)
```

```python
# Part 2. Storing Data 

dict = {} 
for line in url(link):
    line = line.decode().rstrip().lower()
    line = re.sub(r'[^\w\s]', '', line)
    words = line.split(" ") 
    for word in words: 
        if word in dict: 
            dict[word] += 1
        else: 
            dict[word] = 1 

print(dict)

```
