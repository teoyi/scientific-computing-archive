```python
import numpy as np 
import matplotlib.pyplot as plt 
import os
```

```python
v_0 = 100000
v = np.arange(0,200000) 
A = 5
s_v = A * v**2/v_0**2 * np.exp(-v/v_0)
i_0 = A * (v/v_0)*-1 

#plt.plot(v, s_v)
plt.plot(v, i_0)
plt.yscale('log')
plt.xscale('log')
plt.show()

print(i_0)
```

```python
x1 = np.arange(0, 300)
x2 = np.arange(0, 100)
x3 = np.arange(0, 100)
x4 = np.arange(0, 300)
x5 = np.arange(0, 300)
x6 = np.arange(0, 300)

y1 = x1 
y2 = np.exp(-x2/10)
y3 = 2*np.exp(-x3/5)
y4 = x4**4 * np.exp(-x4/10)
y5 = 1/x5 * np.exp(-x5/10)
y6 = 1/(x6+x6**2)

plt.figure() 
plt.loglog(x1,y1, label =r'y = x')
plt.title('a)')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.savefig('a22_a.png')

plt.figure() 
plt.loglog(x2,y2, label = r'$y = e^{-\frac{x}{10}}$')
plt.loglog(x3, y3, label = r'$y = 2e^{-\frac{x}{5}}$')
plt.title('b)')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.savefig('a22_b.png')


plt.figure() 
plt.loglog(x4,y4, label = r'$y = x^4e^{-\frac{x}{10}}$')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.savefig('a22_c.png')


plt.figure()
plt.loglog(x5,y5, label = r'$y = \frac{1}{x}e^{-\frac{x}{10}}$')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.savefig('a22_d.png')


plt.figure()
plt.loglog(x6,y6, label = r'$y = \frac{1}{x+x^2}$')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.savefig('a22_e.png')

plt.show()
```
